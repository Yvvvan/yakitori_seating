```python
# 读取csv
import csv
# 随机打乱座位
import itertools
import random
import copy





for script_no in [5]:
    seating = {}
    
    seed = 20240615
    random.seed(seed)
    
    for _round in range(1, 7): # 6轮
        for _group in range(1, 20): # 19组
            seating[_round, _group] = []
            
    with open('golfer_solution/golfer_solution (%d).csv'%script_no, 'r') as f:
        # 第一行是标题
        reader = csv.reader(f)
        header = next(reader)
        for row in reader:
            # 玩家
            player = int(row[0].split(' ')[1])
            # 第一轮
            round1group = row[1].split(' ')[1]
            seating[1, int(round1group)].append(player)
            # 第二轮
            round2group = row[2].split(' ')[1]
            seating[2, int(round2group)].append(player)
            # 第三轮
            round3group = row[3].split(' ')[1]
            seating[3, int(round3group)].append(player)
            # 第四轮
            round4group = row[4].split(' ')[1]
            seating[4, int(round4group)].append(player)
            # 第五轮
            round5group = row[5].split(' ')[1]
            seating[5, int(round5group)].append(player)
            # 第六轮
            round6group = row[6].split(' ')[1]
            seating[6, int(round6group)].append(player)
            
        # initialize player seating
        player_seating = {}
        player_seating_show = {}
        for _player in range(1, 4*19+1): # 4*19个玩家
            player_seating[_player] = [0, 0, 0, 0] # east, south, west, north
            player_seating_show[_player] = {1:(),2:(),3:(),4:(),5:(),6:()}
            
        # initialize group seating
        seating_copy = copy.deepcopy(seating)

        error = False
        for _round in range(1, 7):
            print()
            for _group in range(1, 20):
                
                p1 = seating_copy[_round, _group][0]
                p2 = seating_copy[_round, _group][1]
                p3 = seating_copy[_round, _group][2]
                p4 = seating_copy[_round, _group][3]
                p1seating = player_seating[p1] #[east,south,west,north]
                p2seating = player_seating[p2]
                p3seating = player_seating[p3]
                p4seating = player_seating[p4]
                
                done = False
                
                for _permutation in itertools.permutations([p1, p2, p3, p4]):
                    east = _permutation[0]
                    south = _permutation[1]
                    west = _permutation[2]
                    north = _permutation[3]
                    
                    ## 往最少的地方坐
                    # if player_seating[east][0] is min(player_seating[east]) 
                    if player_seating[east][0] == min(player_seating[east]) and player_seating[south][1] == min(player_seating[south]) and player_seating[west][2] == min(player_seating[west]) and player_seating[north][3] == min(player_seating[north]):
                        seating_copy[_round, _group] = [east, south, west, north]
                        player_seating[east][0] += 1
                        player_seating[south][1] += 1
                        player_seating[west][2] += 1
                        player_seating[north][3] += 1
                        player_seating_show[east][_round] = (_group,'東')
                        player_seating_show[south][_round] = (_group,'南')
                        player_seating_show[west][_round] = (_group,'西')
                        player_seating_show[north][_round] = (_group,'北')
                        done = True
                        print (f'{east}-{south}-{west}-{north}')
                        break
                    
                if not done:
                    
                    for _permutation in itertools.permutations([p1, p2, p3, p4]):
                        east = _permutation[0]
                        south = _permutation[1]
                        west = _permutation[2]
                        north = _permutation[3]
                        
                        ## 往第二少的地方坐
                        # if player_seating[east][0] in sorted(player_seating[east])[:2]
                        if player_seating[east][0] in sorted(player_seating[east])[:2] and player_seating[south][1] in sorted(player_seating[south])[:2] and player_seating[west][2] in sorted(player_seating[west])[:2] and player_seating[north][3] in sorted(player_seating[north])[:2]:
                            seating_copy[_round, _group] = [east, south, west, north]
                            player_seating[east][0] += 1
                            player_seating[south][1] += 1
                            player_seating[west][2] += 1
                            player_seating[north][3] += 1
                            player_seating_show[east][_round] = (_group,'東')
                            player_seating_show[south][_round] = (_group,'南')
                            player_seating_show[west][_round] = (_group,'西')
                            player_seating_show[north][_round] = (_group,'北')
                            done = True
                            print (f'{east}-{south}-{west}-{north}')
                            break
                        
                    
                    
                    
                    else:
                        ## 随机坐
                        random.shuffle(seating_copy[_round, _group])
                        player_seating[seating_copy[_round, _group][0]][0] += 1
                        player_seating[seating_copy[_round, _group][1]][1] += 1
                        player_seating[seating_copy[_round, _group][2]][2] += 1
                        player_seating[seating_copy[_round, _group][3]][3] += 1
                        player_seating_show[seating_copy[_round, _group][0]][_round] = (_group,'東')
                        player_seating_show[seating_copy[_round, _group][1]][_round] = (_group,'南')
                        player_seating_show[seating_copy[_round, _group][2]][_round] = (_group,'西')
                        player_seating_show[seating_copy[_round, _group][3]][_round] = (_group,'北')
                        print (f'{east}-{south}-{west}-{north}')
                        
        # check if all players are seated in every wind
        print ()
        print ('script %d:'%script_no)
        for p in player_seating:
            if 0 in player_seating[p]:
                print (f'Player {p} does not have every wind, {player_seating[p]}')
```
