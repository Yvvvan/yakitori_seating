```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

def send_email(sender_email, sender_password, receiver_email, subject, body):
    # 创建一个多部分的邮件信息
    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = subject

    # 将邮件正文附加到邮件对象中
    body_html = f'''
        <html>
            <body>
                {body}
            </body>
        </html>
    '''
    msg.attach(MIMEText(body_html, 'html'))

    try:
        # 创建SMTP服务器对象
        server = smtplib.SMTP('smtp.titan.email', 587)  # 这里以Gmail为例
        server.starttls()  # 启用安全传输模式
        server.login(sender_email, sender_password)  # 登录到SMTP服务器
        text = msg.as_string()  # 将邮件对象转换为字符串
        server.sendmail(sender_email, receiver_email, text)  # 发送邮件
        server.quit()  # 断开与SMTP服务器的连接
        print(f'Email sent successfully to {receiver_email}')
    except Exception as e:
        print(f'Failed to send email. Error: {str(e)}')

# 示例调用
# player_name = 'Yufan'
# player_account = 'yufan'
# send_email(
#     sender_email='contact@berlin-mahjong.club',
#     sender_password='', ## Password
#     receiver_email='test@berlin-mahjong.club',
#     subject='Test Email',
#     body= f'''''')
```
